# Runge Kutta 4th Order Method for Solving System of two ODEs
### $\frac{dy}{dx}=f(x, y, z)$ 
### $\frac{dz}{dx}=g(x, y, z)$
## $k_{1} = hf(x_{n}, y_{n}, z_{n})$
## $l_{1} = hg(x_{n}, y_{n}, z_{n})$
## $k_{2} = hf(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{1}}{2}, z_{n} + \frac{l_{1}}{2})$
## $l_{2} = hg(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{1}}{2}, z_{n} + \frac{l_{1}}{2})$
## $k_{3} = hf(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{2}}{2}, z_{n} + \frac{l_{2}}{2})$
## $l_{3} = hg(x_{n} + \frac{h}{2}, y_{n} + \frac{k_{2}}{2}, z_{n} + \frac{l_{2}}{2})$
## $k_{4} = hf(x_{n} + h, y_{n} + k_{3}, z_{n} + l_{3})$
## $l_{4} = hg(x_{n} + h, y_{n} + k_{3}, z_{n} + l_{3})$
## $y_{n+1} = y_{n} + \frac{1}{6}(k_{1} + 2k_{2} + 2k_{3} + k_{4})$ 
## $z_{n+1} = z_{n} + \frac{1}{6}(l_{1} + 2l_{2} + 2l_{3} + l_{4})$ 
## Example
Apply the Runge Kutta 4th Order Method to obtain $y(1)$ and $z(1)$ from the following system of first order ODEs. Use the step size $h=0.05$ 
\begin{equation}
\frac{dy}{dx} = 2y + 4z \hspace{5cm} y(0) = 1
\end{equation}
\begin{equation}
\frac{dz}{dx} = -y + 6z \hspace{5cm} z(0) = 5
\end{equation}

```python
import math
import matplotlib.pyplot as plt
```

```python
#Define the two Functions

#The differential equations are dy/dx = 2y + 4z and dz/dx = -y + 6z with initial conditions y(0)=1 and z(0) = 5 respectively.
f = lambda x,y,z: 2*y + 4*z
g = lambda x,y,z: -y + 6*z   
```

```python
#Initialize Values

#where 'x0 = 0' represents the initial value of x
x0 = 0

#'xf = 1' represents the final value of x
xf = 1

#'h = 0.05' represents the step size
h = 0.05

#'n' represents the number of steps
n = int((xf - x0)/h)

#'x0 = 0','y0 = 1' and 'z0 = 5'represent the initial condition of x, y and z
x0 = 0; y0 = 1; z0 = 5

x = x0
y = y0
z = z0

#Visualization code for initialization

#Header for the table
print('x \t\t y \t\t z')

#The initial values of x, y and f(x)
print('%f \t %f \t %f' %(x, y, z))

#Initialize list for plotting
x_plot = []; y_RK4 = []; z_RK4 = []

#Runge Kutta 4th order method implementation
for i in range (1, n+1):
    x_plot.append(x)
    y_RK4.append(y)
    z_RK4.append(z)
    
    #Calculating the system of first order ODEs using RK4 method
    k1 = h * f(x,y,z)
    l1 = h * g(x,y,z)  
    k2 = h * f(x + h/2, y + k1/2, z + l1/2)
    l2 = h * g(x + h/2, y + k1/2, z + l1/2)  
    k3 = h * f(x + h/2, y + k2/2, z + l2/2)
    l3 = h * g(x + h/2, y + k2/2, z + l2/2)  
    k4 = h * f(x + h, y + k3, z + l3)
    l4 = h * g(x + h, y + k3, z + l3)  
    #Approximate solution of y and z
    y = y + 1/6 * (k1 + 2*k2 + 2*k3 + k4)
    z = z + 1/6 * (l1 + 2*l2 + 2*l3 + l4)
    #Increase x by step size to calculate the next y estimate
    x = x + h
    
    print('%f \t %f \t %f' %(x, y, z))
    
#Visualization codes for plotting the approximate and analytical solution of the differential equation
x_plot.append(x)
y_RK4.append(y)
z_RK4.append(z)
plt.plot(x_plot,y_RK4, 'rs--')
plt.plot(x_plot, z_RK4, 'g^--')          
plt.xlabel('$x$', fontsize=20)
plt.ylabel('$y, z$', fontsize=20)
plt.legend(["$y' = 2y + 4z$", "$z' = -y + 6z$"])
plt.title('RK4 Method for System of two First Order ODEs', fontsize=20)
plt.grid(True)
```
