# Runge Kutta 4th Order Method for Solving a System of three ODEs
### $\frac{dx}{dt}=f(t, x, y, z)$ 
### $\frac{dy}{dt}=g(t, x, y, z)$
### $\frac{dz}{dt}=j(t, x, y, z)$
## $k_{1} = hf(t_{n}, x_{n}, y_{n}, z_{n})$
## $l_{1} = hg(t_{n}, x_{n}, y_{n}, z_{n})$
## $m_{1} = hj(t_{n}, x_{n}, y_{n}, z_{n})$
## $k_{2} = hf(t_{n} + \frac{h}{2}, x_{n} + \frac{k_{1}}{2}, y_{n} + \frac{l_{1}}{2}, z_{n} + \frac{m_{1}}{2})$
## $l_{2} = hg(t_{n} + \frac{h}{2}, x_{n} + \frac{k_{1}}{2}, y_{n} + \frac{l_{1}}{2}, z_{n} + \frac{m_{1}}{2})$
## $m_{2} = hj(t_{n} + \frac{h}{2}, x_{n} + \frac{k_{1}}{2}, y_{n} + \frac{l_{1}}{2}, z_{n} + \frac{m_{1}}{2})$
## $k_{3} = hf(t_{n} + \frac{h}{2}, x_{n} + \frac{k_{2}}{2}, y_{n} + \frac{l_{2}}{2}, z_{n} + \frac{m_{2}}{2})$
## $l_{3} = hg(t_{n} + \frac{h}{2}, x_{n} + \frac{k_{2}}{2}, y_{n} + \frac{l_{2}}{2}, z_{n} + \frac{m_{2}}{2})$
## $m_{3} = hj(t_{n} + \frac{h}{2}, x_{n} + \frac{k_{2}}{2}, y_{n} + \frac{l_{2}}{2}, z_{n} + \frac{m_{2}}{2})$
## $k_{4} = hf(t_{n} + h, x_{n} + k_{3}, y_{n} + l_{3}, z_{n} + m_{3})$
## $l_{4} = hg(t_{n} + h, x_{n} + k_{3}, y_{n} + l_{3}, z_{n} + m_{3})$
## $m_{4} = hj(t_{n} + h, x_{n} + k_{3}, y_{n} + l_{3}, z_{n} + m_{3})$
## $x_{n+1} = x_{n} + \frac{1}{6}(k_{1} + 2k_{2} + 2k_{3} + k_{4})$ 
## $y_{n+1} = y_{n} + \frac{1}{6}(l_{1} + 2l_{2} + 2l_{3} + l_{4})$ 
## $z_{n+1} = z_{n} + \frac{1}{6}(m_{1} + 2m_{2} + 2m_{3} + m_{4})$ 

## Example
Apply the Runge Kutta 4th Order Method to obtain $x(20)$, $y(20)$ and $z(20)$ from the Lorenz Equations. Use the step size $h=0.03125$ 
\begin{equation}
\frac{dx}{dt} = \sigma y - \sigma x \hspace{5cm} x(0) = 1
\end{equation}
\begin{equation}
\frac{dy}{dt} = rx - y -xz \hspace{4.5cm} y(0) = 1
\end{equation}
\begin{equation}
\frac{dz}{dt} = xy - bz \hspace{5cm} z(0) = 1
\end{equation}

```python
import math
import matplotlib.pyplot as plt
```

```python
#Define the three Functions

#The Lorenz Equations are dx/dt = sigma (y - x), dy/dt = rx - y - xz and dz/dt = xy - bz with initial conditions x(0)=1, y(0)=1 and z(0) = 1 respectively.
f = lambda t,x,y,z: math.sigma*y - math.sigma*x
g = lambda t,x,y,z: r*x - y - x*z
j = lambda t,x,y,z: x*y - b*z
```

```python
#Initialize Values

#where 't0 = 0' represents the initial value of t
t0 = 0

#'tf = 20' represents the final value of t
tf = 20

#'h = 0.03125' represents the step size
h = 0.03125

#'n' represents the number of steps
n = int((tf - t0)/h)

#'t0 = 0, 'x0 = 1','y0 = 1' and 'z0 = 1'represent the initial condition of t, x, y and z
t0 = 0; x0 = 1; y0 = 1; z0 = 1

t = t0
x = x0
y = y0
z = z0

# The parameters of the Lorenz Equations are
math.sigma = 10
b = 8/3
r = 28

#Visualization code for initialization

#Header for the table
print('t \t\t x \t\t y \t\t z')

#The initial values of x, y and f(x)
print('%f \t %f \t %f \t %f' %(t, x, y, z))

#Initialize list for plotting
t_plot = []; x_RK4 = []; y_RK4 = []; z_RK4 = []

#Runge Kutta 4th order method implementation
for i in range (1, n+1):
    t_plot.append(t)
    x_RK4.append(x)
    y_RK4.append(y)
    z_RK4.append(z)
    
    #Calculating the system of first order ODEs using RK4 method
    k1 = h * f(t,x,y,z)
    l1 = h * g(t,x,y,z) 
    m1 = h * j(t,x,y,z)
    k2 = h * f(t + h/2, x + k1/2, y + l1/2, z + m1/2)
    l2 = h * g(t + h/2, x + k1/2, y + l1/2, z + m1/2)
    m2 = h * j(t + h/2, x + k1/2, y + l1/2, z + m1/2)
    k3 = h * f(t + h/2, x + k2/2, y + l2/2, z + m2/2)
    l3 = h * g(t + h/2, x + k2/2, y + l2/2, z + m2/2)  
    m3 = h * j(t + h/2, x + k2/2, y + l2/2, z + m2/2)
    k4 = h * f(t + h, x + k3, y + l3, z + m3)
    l4 = h * g(t + h, x + k3, y + l3, z + m3)  
    m4 = h * j(t + h, x + k3, y + l3, z + m3) 
    #Approximate solution of x, y and z
    x = x + 1/6 * (k1 + 2*k2 + 2*k3 + k4)
    y = y + 1/6 * (l1 + 2*l2 + 2*l3 + l4)
    z = z + 1/6 * (m1 + 2*m2 + 2*m3 + m4)
    #Increase t by step size to calculate the next x, y and z estimate
    t = t + h
    
    print('%f \t %f \t %f \t %f' %(t, x, y, z))
    
#Visualization codes for plotting the approximate solution of the Lorenz Equation
t_plot.append(t)
x_RK4.append(x)
y_RK4.append(y)
z_RK4.append(z)
plt.plot(t_plot, x_RK4, 'r--')
plt.plot(t_plot, y_RK4, 'g--')
plt.plot(t_plot, z_RK4, 'b--')
plt.xlabel('$Time$', fontsize=20)
plt.ylabel('$Quantity$', fontsize=20)
plt.legend(["$x' = \sigma y - \sigma x$", "$y' = rx - y - xz$", "$z' = xy - bz$"])
plt.title('RK4 Method for Solving the Lorenz Equation', fontsize=20)
```
